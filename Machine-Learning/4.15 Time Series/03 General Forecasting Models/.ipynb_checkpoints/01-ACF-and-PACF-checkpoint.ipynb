Before we can investigate <em>autoregression</em> as a modeling tool, we need to look at <em>covariance</em> and <em>correlation</em> as they relate to lagged (shifted) samples of a time series.

**What is Correlation ?**

Correlation is a measure of the strength of the linear relationship between two variables.<br>
- Ranges from -1 to +1
- Closer to +1 the stronger the +ve linear relationship
- Closer to -1 the stronger the -ve linear relationship
- Closer to 0, the weaker the linear relationship or association

**Covariance vs Correlation**
- A measure used to indicate the extent to which two random variables change in tandem is known as covariance. A measure used to represent how strongly two random variables are related known as correlation.
- Covariance is nothing but a measure of correlation. On the contrary, correlation refers to the scaled form of covariance.
- The value of correlation takes place between -1 and +1. Conversely, the value of covariance lies between -∞ and +∞.
- Covariance is affected by the change in scale, i.e. if all the value of one variable is multiplied by a constant and all the value of another variable are multiplied, by a similar or different constant, then the covariance is changed. As against this, correlation is not influenced by the change in scale.
- Correlation is dimensionless, i.e. it is a unit-free measure of the relationship between variables. Unlike covariance, where the value is obtained by the product of the units of the two variables.

Correlation is a special case of covariance which can be obtained when the data is standardized. 

Now, when it comes to making a choice, which is a better measure of the relationship between two variables: 
- **correlation is preferred over covariance, because it remains unaffected by the change in location and scale**
- and can also be used to make a comparison between two pairs of variables.

**What is Auto-Correlation ?**

Auto-correlation is the correlation of the series with itself lagged by 'x' time units. 

This is what an auto-correlation plot shows. So, the y-axis is the correlation and the x-axis is the number of time units of lag.

**Why we do this auto-correlation ?**
- Imagine we have some sales data. 
- We can plot an auto-correlation plot on that
- That plot will tell us, how correlated are today's sales to yesterday's sale ? Is it going up or down or stationary.

**What is ACF (Auto Correlation Function) ?**

The ACF describes the auto-correlation between observation and another observation at a prior time step, that includes direct and in-direct dependence information.

**What is PACF (Partial-ACF)?**

The PACF only describes the direct relationship between an observation and its lag.

**These two plots can help choose order parameters for ARIMA based models.**
But, it is usually much easier to perform a grid search of the parameter values, rather than attempt to read these plots directly.

## Autocorrelation for 1-Dimension

Function for autocorrelation - acf(pass a series here)

```python
df = pd.DataFrame({'a':[13,5,11,12,9]})
```

```python
arr3 = acf(df['a'])
arr3
```

## Partial Autocorrelation

pacf_yw = partial auto-correlation yule walker method
- pass a Series, 
- pass for specific number of lags (nlags = number of rows -1 )
- pass the method 'mle' (Maximum Likelihood Estimation) in order to use the biased ACF coefficients. Alternatively you could also pass method='unbiased' which is the default

```python
arr4 = pacf_yw(df['a'],nlags=4,method='mle')
arr4
```

<div class="alert alert-info"><strong>NOTE:</strong> We passed in <tt><font color=black>method='mle'</font></tt> above in order to use biased ACF coefficients. "mle" stands for "maximum likelihood estimation". Alternatively we can pass <tt>method='unbiased'</tt> (the statsmodels default):</div>

```python
arr5 = pacf_yw(df['a'],nlags=4,method='unbiased')
arr5
```

### Partial Autocorrelation with OLS
This provides partial autocorrelations with <a href='https://en.wikipedia.org/wiki/Ordinary_least_squares'>ordinary least squares</a> (OLS) estimates for each lag instead of Yule-Walker.

```python
arr6 = pacf_ols(df['a'],nlags=4)
arr6
```

**The Basic premise behind why there are different correllation functions is that there are different ways of calculating the correlation itself**

One of the common method is Pearson's correlation method

# Plotting
The arrays returned by <tt>.acf(df)</tt> and <tt>.pacf_yw(df)</tt> show the magnitude of the autocorrelation for a given $y$ at time $t$. Before we look at plotting arrays, let's look at the data itself for evidence of autocorrelation.

Pandas has a built-in plotting function that plots increasing $y_t$ values on the horizontal axis against lagged versions of the values $y_{t+1}$ on the vertical axis. If a dataset is non-stationary with an upward trend, then neighboring values should trend in the same way. Let's look at the <strong>Airline Passengers</strong> dataset first.

```python
from pandas.plotting import lag_plot
```

```python
lag_plot(df1['Thousands of Passengers']);
```

Visually this shows evidence of a very strong autocorrelation; as $y_t$ values increase, nearby (lagged) values also increase.

Now let's look at the stationary <strong>Daily Total Female Births</strong> dataset:

```python
lag_plot(df2['Births']);
```

As expected, there is little evidence of autocorrelation here.

## ACF Plots
Plotting the magnitude of the autocorrelations over the first few (20-40) lags can say a lot about a time series.

**ACF Plots on Non-stationary data**

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
df1.plot()
#This has clear seasonality and trend
#SO this is definitly non-stationary
```

Lets Plot ACF for df1

```python
plot_acf(df1, lags=40);
#By default it will do as many lags as there are steps in series
#But often, we do not need that many
#If we look at df1, we have 144 observarions, so 40 lags will be fine
```

**Notice:**
- There is a clear indication of seasonality in the data, becuase there are these general peaks in the auto-correlation.
- If you notice, there is a general trend of increase on yearly basis, which happens around the 12th mark, 24th mark etc in the graph

**Now what does this shaded region actually means ?**
- This shaded region, by default, this is 95% confidence interval
- Basically what it is suggesting is the correlation values outside this 95% confidence interval, are very highly likely to be a correlation or correlated.

**ACF Plots on tationary data**

```python
# Now let's plot the autocorrelation at different lags
title = 'Autocorrelation: Daily Female Births'
lags = 40
plot_acf(df2,title=title,lags=lags);
```

This is a typical ACF plot for stationary data, with lags on the horizontal axis and correlations on the vertical axis. The first value $y_0$ is always 1. A sharp dropoff indicates that there is no AR component in the ARIMA model.

## PACF Plots
Partial autocorrelations work best with stationary data. Let's look first at <strong>Daily Total Female Births</strong>:

```python
title='Partial Autocorrelation: Daily Female Births'
lags=40
plot_pacf(df2,title=title,lags=lags);
```

To make the <strong>Airline Passengers</strong> data stationary, we'll first apply <em>differencing</em>:

```python
from statsmodels.tsa.statespace.tools import diff

df1['d1'] = diff(df1['Thousands of Passengers'],k_diff=1)
df1['d1'].plot(figsize=(12,5));
```

```python
title='PACF: Airline Passengers First Difference'
lags=40
plot_pacf(df1['d1'].dropna(),title=title,lags=np.arange(lags));  # be sure to add .dropna() here!
```
